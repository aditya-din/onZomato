## Feature Selection using kbest.


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression #regression because all x and y will be countinuous.
```

```python
#selecting x(features) and y for prediction.
x = df.drop(columns = ['url','address','name','rate','phone','location','rest_type','dish_liked','cuisines','reviews_list','menu_item','listed_in(city)'])
y=df.rate
select = SelectKBest(score_func = f_regression,k=10) #finding best 10 columns which will effect rate.
fit = select.fit(x,y)
dfScores = pd.DataFrame(fit.scores_,columns=['scores'])
dfColumns = pd.DataFrame(x.columns,columns=['columns'])
dfNew = pd.concat([dfColumns,dfScores],axis=1)
print("\nThese are the columns which are effecting the 'rate' most. Higher the score - Higher the effect.")
dfNew.sort_values(['scores'],ascending=False)
```
